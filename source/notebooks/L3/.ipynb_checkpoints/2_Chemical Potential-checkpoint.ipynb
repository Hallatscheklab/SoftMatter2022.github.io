## Chemical Potential and Osmotic Pressure

So far we have considered either isolated systems, meaning that no energy nor particle exchange is possible, and closed systems, which allow an exchange of energy with a reservoir (bath). We will now have a look at a system that also allows an exchange of particles with a reservoir. For these systems, we will define the **chemical potential** $\mu$ which we will identify as the **free energy change when adding one molecule of a given species to a thermodynamic system** at constant $p$ and $T$.

We consider a system of two components A and B in contact with a heat bath and a diffusive equilibrium between A and B. 

![Chemical Potential](img/chem_pot.png)

The total free energy of our system is then defined as

\begin{equation}
G=G_{\mathrm A}+G_{\mathrm B}.
\end{equation}

The total number of particles in the system is also conserved and given by $N_{\rm tot}=N_A+N_B$ In equilibrium we require the free energy of the system to be a minimum with respect to the particle number such that

\begin{equation}\frac{\mathrm dG}{\mathrm dN}=\frac{\mathrm dG_{\mathrm A}}{\mathrm dN_{\mathrm A}}|_{T,p}+\frac{\mathrm dG_{\mathrm B}}{\mathrm dN_{\mathrm B}}|_{T,p}=0.\end{equation}

Since the total number of particles is conserved, we also find in equilibrium that from which follows that

\begin{equation}
\frac{\mathrm dG_{\mathrm A}}{\mathrm dN_{\mathrm A}}=-\frac{\mathrm dG_{\mathrm B}}{\mathrm dN_{\mathrm B}}
\end{equation}

and with $dN_A=-dN_B$ also 

\begin{equation}
\frac{\mathrm dG_{\mathrm A}}{\mathrm dN_{\mathrm A}}=\frac{\mathrm dG_{\mathrm B}}{\mathrm dN_{\mathrm A}}.
\end{equation}

Apparently, in equilibrium the change in the free energy of system $A$ needs to be the same (or negative) of system $B$.

We can now define a new quantity

\begin{equation}
\mu(T,p,N)=\frac{\mathrm dG}{\mathrm dN}|_{T,P}
\end{equation}

which is termed the chemical potential, which is the change in the free energy when adding a particle to the system. The chemical potential may be interpreted as the capacity of a system to change the number of particles. We will have a closer look at the chemical potential with an example now. The previous condition states that the chemical potential of the two systems in the same in equilibrium.

**Free energy of a dilute solution**

We would like to calculate the free energy and the chemical potential for a dilute solution of some _solutes_ in a solvent, which we just term H$_2$O. Actually, no additional information on the details of the solute and solvent are currently required, though if we want to have numbers, we would need to know which solute or solvent we are looking at.

![Solute Solvent](img/solute_solvent.png)    

    
Suppose we have
    
- $N_{\rm H_2 O}$ solvent molecules (e.g. water)
- $N_s$ solute molecules (e.g. proteins)
    
    
The solute chemical potential is defined by

\begin{equation}
\mu_{\rm solute}=\left( \frac{\partial G_{\rm tot}}{\partial N_{\rm s}}\right)|_{p,T}
\end{equation}

or intuitively as

\begin{equation}
\mu_{\rm solute}=G_{\rm tot}(N_{\rm s}+1)-G_{\rm tot}(N_{\rm s}).
\end{equation}

The total free energy $G_{\rm tot}$ consists of the enthalpy of forming the solven molecules $N_{\rm{H_{2}O}}\varepsilon_{\rm H_{2}O}$ and the enthalpy solvating the solute proteins $N_{\rm s}\varepsilon_{\rm s}$. 

\begin{equation}
G_{\rm tot}=N_{\rm{H_{2}O}}\varepsilon_{\rm H_{2}O}+N_{\rm s}\varepsilon_{\rm s}-TS_{\rm mix}.
\end{equation}

The last term $TS_{\rm mix}$ denotes the entropy for mixing the solute and the solvent molecules.     
In the following, we are interested in the last term, which is the entropic contribution to the free energy containing the entropy of mixing solute and solvent.

The mixing entropy can be calculated by either the Gibbs definition

\begin{equation}
S_{\rm mix}=-k_{\rm B}\sum_{i}p_{i}\ln(p_{i})
\end{equation}

or the Boltzmann definition

\begin{equation}
S_{\rm mix}=k_{\rm B}\ln(W).
\end{equation}

We will use the latter definition including the number of possible configurations to obtain the mixing entropy. If the total number of molecules in the volume is $N=N_{\rm H_2O}+N_s$ we can write down the number of different ways to arrange the molecules as previously done, we obtain
    
\begin{equation}
W(N_{\rm H_2O}, N_s )=\frac{N!}{N_{\rm H_2O}! N_s!}
\end{equation}
    

As in our example with the DNA binding, we can apply the Stirling formula for large $N$ and find    
    
\begin{equation}
S_{\rm mix}=-k_B \left ( N_{\rm H_2O} \ln\left (\frac{N_{\rm H_2O}}{N_{\rm H_2O}+N_s}\right) +N_s\ln\left (\frac{N_s}{N_{\rm H_2O}+N_s} \right )\right )
\end{equation}    

To really go in the dilute limit, the number of solute molecules should be much larger than the number of solvent molecules, i.e.    
    
$$\frac{N_s}{N_{\rm H_2O}}\ll 1$$
    
which allows us to approximate the solution with
    
\begin{equation}
S_{\rm mix}\approx -k_B \left ( N_{\rm H_2O} \ln \left ( 1-\frac{N_s}{N_{\rm H_2O}}\right) +N_s\ln\left ( \frac{N_s}{N_{\rm H_2O}}\right)\right )
\end{equation}
    
Using 
    
$$\ln(1+\epsilon)\approx \epsilon$$

we may finally write

\begin{equation}
S_{\rm mix}\approx -k_B \left ( N_{s} \ln \left ( \frac{N_s}{N_{\rm H_2O}}\right) -N_s\right )
\end{equation}
    
and the total free energy is then given by
    
\begin{equation}
G_{\rm tot}=N_{\rm{H_{2}O}}\varepsilon_{\rm H_{2}O}+N_{\rm s}\varepsilon_{\rm s}--k_B T \left ( N_{s} \ln \left ( \frac{N_s}{N_{\rm H_2O}}\right) -N_s\right )
\end{equation}

We can now come back and calculate the chemical potential of the solute as noted above 
    
    
\begin{equation}
\mu_s=\epsilon_s+k_B T \ln \left ( \frac{N_s}{N_{\rm H_2O}}\right )
\end{equation}
    
As we rather work with concentrations than number of molecules we can use $c=N_s/V$ and $c_0=c_{\rm H_2O}=N_{\rm H_2O}/V$  to write
    
\begin{equation}
\mu_s=\epsilon_s+k_B T \ln \left ( \frac{c}{c_0}\right )
\end{equation}    
    
    
The value of $c_0$ thereby acts as a reference point, which is commonly chosen to be at $c_0=1\,M$. The value of $\epsilon_s$ is termed the standard chemical potential $\mu_s0$.
The standard chemical potential is measured at standard thermodynamic conditions, i.e. $p_0=101.3\, {\rm kPa}$ and $T=293.15\, {\rm K}$ and $c_0=1\, {\rm M}$.

**Chemical Potential**
    
The chemical potential of a component $i$ consists of two components
    
    
    
\begin{equation}
\mu_i=\mu_i0+k_B T \ln \left ( \frac{c_i}{c_i0}\right )
\end{equation}
    
- The standard chemical potential is a molecular property (see slides) that contains 
    - the internal energy/enthalpy to create the molecule, 
    - the conformational entropy of the molecule and  
    - the enthalpy and entropy contributions when bringing the molecule into contact with the solvent (solvation free energy. 
- The logarithm term     $k_B T \ln \left ( \frac{c_i}{c_i0}\right )$ is the pure mixing/dilution entropy reflecting the different abundance of solute and solvent due to the resulting numbers of microstates
